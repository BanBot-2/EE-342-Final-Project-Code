```python
import pandas as pd
import os
```

```python
# 1. Step UP one folder out of 'notebooks', then INTO 'data/raw'
data_path = '../data/raw'
```

```python
# 2. Load the pre-calculated features (x)
x_train = pd.read_csv(
    os.path.join(data_path, 'train', 'X_train.txt'),
    sep=r'\s+', 
    header=None
)

# 3. Load the activity labels (y)
y_train = pd.read_csv(
    os.path.join(data_path, 'train', 'y_train.txt'), 
    sep=r'\s+', 
    header=None, 
    names=['Activity_Label']
)

# 4. Load the subject identifiers
subject_train = pd.read_csv(
    os.path.join(data_path, 'train', 'subject_train.txt'), 
    sep=r'\s+', 
    header=None, 
    names=['Subject_ID']
)
```

```python
# 5. Combine everything into a single DataFrame
train_df = pd.concat([subject_train, y_train, x_train], axis=1)

# Display the first few rows and the shape
print(f"Dataframe shape: {train_df.shape}")
print(train_df.head())
```

```python
# 6. Load the text labels 
activity_labels = pd.read_csv(
    os.path.join(data_path, 'activity_labels.txt'), 
    sep=r'\s+', 
    header=None, 
    names=['Activity_Label', 'Activity_Name']
)

# 7. Merge the text names into the main training dataframe
train_df = train_df.merge(activity_labels, on='Activity_Label', how='left')

# Move the Activity_Name column to the front so it is easy to read
cols = train_df.columns.tolist()
cols.insert(2, cols.pop(cols.index('Activity_Name')))
train_df = train_df.reindex(columns=cols)

# Verify the mapping worked (Label 5 should now show as STANDING)
print(train_df[['Subject_ID', 'Activity_Label', 'Activity_Name']].head())
```

```python
# 8. Load the feature names from features.txt
features = pd.read_csv(
    os.path.join(data_path, 'features.txt'), 
    sep=r'\s+', 
    header=None, 
    names=['Feature_Index', 'Feature_Name']
)

# 9. Extract the names into a Python list
feature_names = features['Feature_Name'].tolist()

# 10. Create a dictionary that maps the number (0-560) to the text name
# Example: {0: 'tBodyAcc-mean()-X', 1: 'tBodyAcc-mean()-Y', ...}
rename_mapping = {i: name for i, name in enumerate(feature_names)}

# 11. Rename the columns in your dataframe
train_df = train_df.rename(columns=rename_mapping)

# Display the final, fully labeled dataset!
print(train_df.head())
```
