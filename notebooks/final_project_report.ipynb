## Setup

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn.functional as F
```

```python
data_path = '../data/raw/train' # path to raw signals within dataset
```

## Loading Training Data

```python
# Loads raw acceleration data (which includes gravity).
accel_x = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/total_acc_x_train.txt'),
    sep=r'\s+',
    header=None
).values

accel_y = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/total_acc_y_train.txt'),
    sep=r'\s+',
    header=None
).values

accel_z = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/total_acc_z_train.txt'),
    sep=r'\s+',
    header=None
).values

# Stacks into a 3D NumPy array of shape: (samples, time_steps, channels).
accel_array = np.stack([accel_x, accel_y, accel_z], axis=-1)

# Converts to a PyTorch tensor (dtype=torch.float32 is standard for training inputs).
accel_tensor = torch.tensor(accel_array, dtype=torch.float32)

print(f"Raw accelerometer tensor shape: {accel_tensor.shape}")
```

```python
# Loads raw gyroscope data.
gyro_x = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/body_gyro_x_train.txt'),
    sep=r'\s+',
    header=None
).values

gyro_y = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/body_gyro_y_train.txt'),
    sep=r'\s+',
    header=None
).values

gyro_z = pd.read_csv(
    os.path.join(data_path, 'Inertial Signals/body_gyro_z_train.txt'),
    sep=r'\s+',
    header=None
).values

# Stacks into a 3D NumPy array of shape: (samples, time_steps, channels).
gyro_array = np.stack([gyro_x, gyro_y, gyro_z], axis=-1)

# Converts to a PyTorch tensor (dtype=torch.float32 is standard for training inputs).
gyro_tensor = torch.tensor(gyro_array, dtype=torch.float32)

print(f"Raw gyro tensor shape: {gyro_tensor.shape}")
```

```python
# Loads raw labels for the gyro and accel data.
y_train_raw = pd.read_csv(
    os.path.join(data_path, 'y_train.txt'),
    sep=r'\s+',
    header=None
).values

# Flattens labels down to a 1D NumPy array and subtracts 1 to make labels go from 0-5.
y_train_flattened = y_train_raw.squeeze() - 1

# Converts to a 1D PyTorch tensor of integers (dtype=torch.long is required for labels). 
y_tensor = torch.tensor(y_train_flattened, dtype=torch.long)

print(f"Labels tensor shape: {y_tensor.shape}")
print(f"Unique labels in tensor: {torch.unique(y_tensor)}")
```
